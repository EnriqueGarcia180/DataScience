## 2. Operaciones

### Operaciones aritméticas

Con Python podemos realizar operaciones aritméticas como suma, resta , multiplicación, etc. 
- **Suma**: se realiza con el operador **+**
- **Resta**: se realiza con el operador **-**
- **Multiplicación**: se realiza con el operador *
- **División**: se realiza con el operador **/**
- **Residuo**: obtenemos el residuo de una división mediante el operador **%**
- **Potencia**: se realiza con el operador **

```python
# Suma
print("Sumar 2 más 2 nos da: ", 2 + 2)
# Resta
print("Restar 10 menos 8 nos da: ", 10 - 8)
# Multiplicación
print("Multiplicar 24 por 3 nos da: ", 24 * 3)
# División
print("Dividir 50 entre 6 nos da: ", round(50 / 6, 2))  
# Residuo
print("El residuo de la operación anterior es: ", 50 % 6)
# Potencia
print("2 elevado a la 8 nos da: ", 2**8)
```

### Operaciones de comparación

En Python, utilizamos los operadores de comparación para comparar entre dos o más valores. 

Como resultado, siempre obtendremos un valor de tipo booleano, es decir, **verdadero** o **falso**.

##### - Igual
Se realiza mediante el operador **==**. True si el operando de la izquierda es igual que el de derecha, False en caso contrario.
##### - Distinto
Se realiza mediante el operador **!=**. True si los operandos son distintos; False en caso contrario.
##### - Mayor que
Se realiza mediante el operador **>**. True si el operando de la izquierda es estrictamente mayor que el de la derecha, False en caso contrario.
##### - Mayor o igual que
Se realiza mediante el operador **>=**. True si el operando de la izquierda es mayor o igual que el de la derecha; False en caso contrario.
##### - Menor que
Se realiza mediante el operador **<**. True si el operando de la izquierda es estrictamente menor que el de la derecha; False en caso contrario.
##### - Menor o igual que
Se realiza mediante el operador **<=**. True si el operando de la izquierda es menor o igual que el de la derecha; False en caso contrario.

```python
primer_dato = 40
segundo_dato = 51 

# Igual
print('¿El primer dato es igual al segundo?:', primer_dato == segundo_dato)
# Distino
print('¿El primer dato es distinto al segundo?:', primer_dato != segundo_dato)
# Mayor que / Mayor o igual que
print('¿El primer dato es mayor o igual al segundo?:', primer_dato >= segundo_dato)
# Menor que / Menor o igual que
print('¿El primer dato es menor o igual al segundo?:', primer_dato <= segundo_dato)
```

### Operaciones lógicas

Los operadores lógicos comparan declaraciones y devuelven un valor de True o False, estos también se conocen como operadores booleanos.

Se utiliza un operador lógico para tomar una decisión basada en múltiples condiciones.

- **Not:** obtenemos lo opuesto (la negación) a una declaración dada.
- **And:** revisa si ambas declaraciones son verdaderas.
- **Or:** revisa si al menos una de las declaraciones son verdaderas.

**Nota**. Las evaluaciones lógicas se ejecutarán en el orden previamente presentado (primero Not, luego And y finalmente Or).


<p><img src="https://rea.ceibal.edu.uy/elp/logica-para-informatica/TablasFinales.png" width="450"</p>

```python
# Ejecutar todas las combinaciones posibles con el operador Not
print("Operador not: ", not True, not False)
# Ejecutar todas las combinaciones posibles con el operador And
print("Operador and: ", True and True, True and False, False and True, False and False)
# Ejecutar todas las combinaciones posibles con el operador Or
print("Operador or: ", True or True, True or False, False or True, False or False)
```

```python
print(not 20 < 10 )
print(10 > 5 and 5 == 5)
print(10 != 0 or 3 < 1)
```

Podemos realizar evaluaciones con operadores lógicos combinados con operadores de comparación 

```python
# Realizar operación lógica a partir de una comparación
print (24 + 2 > 12 + 6 and 10 < 4)
```

La jerarquía previamente mencionada es importante a la hora de realizar una evaluación lógica con más de una declaración.

```python
# Realizar una evaluación lógica considerando más de una declaración
print(True and not True or True)    # esto es un comentario

# Orden o jerarquia: NAO [Not, And, Or]
# Paso 1: True and False or True
# Paso 2: False or True
# Paso 3: True
```

### Operaciones de identidad

Nos indican si dos variables hacen referencia al mismo objeto.

Si dos variables distintas tienen el mismo id(), el resultado de aplicar el operador **is** sobre ellas será True.

- **is:** Nos devuelve **True** si hacen referencia a el mismo objeto.
- **is not:** Nos devuelve **False** si no hacen referencia a el mismo objeto.

```python
# operación de identidad con un integer
x = 35
y = 35

print(id(x))
print(id(y))

print(x is y)
print(x == y)

print(x is not y)
```

```python
# Operación de identidad con una lista
x = [6,10,15]
y = [6,10,15]

print(id(x))
print(id(y))
print(x is y)
print(x == y)
print(x is not y)

```

Note que en el ejemplo anterior, el resultado *(x is y)* es **False** a pesar que las variables contienen los mismos datos. 

Esto es porque al ser una lista un objeto **mutable** (que puede cambiar una vez creada), Python le asignará un **id** distinto sin importar que las variables contengan al principio los mismos datos.

Objeto **MUTABLE**: es un Objeto que Puede cambiar su Valor una vez creada i.e. Las **Listas**. Un Objeto NO-MUTABLE no puede cambiar el valor i.e. constante (35)

## 3. Listas, tuplas y diccionarios

### Listas

Una lista es una “secuencia” de datos ya sean números, cadenas de texto, otras listas, etc. Se definirán entre corchetes **[ ]**, y a partir de la definición de una lista, podemos acceder y extraer un elemento de esta.

```python
#Crear una lista
Alumnos = ["Rodrigo", "José", "Enrique", 0, [1, 'dos', 3]]
print(Alumnos)

print(Alumnos[0])
print(Alumnos[4][1])
```

Podemos acceder a los elementos de la lista mediante su índice (se cuentan posiciones a partir del índice 0 de izquierda a derecha). 

Para acceder al útlimo elemento (o cualquier elemento de derecha a izquierda), comenzamos a contar a partir del -1 y así subsecuentemente.

```python
Alumnos = ["Rodrigo", "José", "Enrique", 0, [1, 'dos', 3]]

#Acceder al primer elemento de la lista
print(Alumnos[0])
print(Alumnos[1])

#Acceder al último elemento de la lista
print(Alumnos[-1])
print(Alumnos[-2])

print(Alumnos[10]) #debe dar Error
```

Las listas son objetos **mutables**, es decir, pueden ser modificados una vez creados. Es posible reemplazar, eliminar y agregar datos a la lista.

```python
Alumnos = ["Rodrigo", "José", "Enrique", 0, [1, 'dos', 3]]
print(Alumnos)

#Reemplazar un elemento de la lista:
Alumnos[2] = "María"
print(Alumnos)

#Eliminar un elemento a la lista
del Alumnos[1] #borra a jose
print(Alumnos)

# reemplazar
Alumnos[-1] = "nuevo"
print(Alumnos)
```

Podemos agregar un elemento a una lista a través del método **append()** y agregar varios elementos a una lista a través del método **extend()**.

```python
Alumnos = ["Rodrigo", "José", "Enrique", 0, [1, 'dos', 3]]

#Agregar un elemento a la lista
Alumnos.append('Carmen')

#Agregar varios elementos a una lista
Alumnos.extend(['Jorge','Ana','Ana'])

print(Alumnos)
```

```python
mi_lista = []

mi_lista.append('Elemento_1')

print(mi_lista)
```

**SLICING**

Podemos acceder a un rango de elementos de un objeto (como las listas) mediante **slicing**. Esta técnica puede interpretarse como realizar un corte al objeto para acceder a una parte de este.

```python
print(Alumnos)

# SLICING !!!

#Diferentes ejemplos de slicing
print('1.-', Alumnos[1:3])

print('2.-', Alumnos[1:])

print('3.-', Alumnos[:2])

print('4.-', Alumnos[-2:])

print('5.-', Alumnos[:-4])
```

Para conocer la posición que tiene un elemento dentro de la lista, utilizamo el método **index()**.

```python
#Conocer la ubicación de un elemento dentro del objeto
Alumnos.index('Ana')
```

Por último, la función **len()** nos permitirá saber cuántos elementos conforman una lista.

```python
#Contar los elementos dentro de una lista
print(len(Alumnos))
print(len(Alumnos[4]))
```

```python
mi_lista = ['Marco', 'Marco']

print(mi_lista.index('Marco'))
```

### Tuples

Son otro tipo de objeto que almacena una serie de valores (secuencia). A diferencia de las listas, las tuples son **inmutables**, es decir, no se pueden modificar (eliminar ni agregar elementos). Las tuples definirán entre paréntesis **()**; son el objeto por default de python para almacenar una secuencia de valores. 

```python
# Crear una tupla
mi_tupla = (5,10,15, (1,2,3), ['a', 'b'])
print(mi_tupla)
```

Es posible convertir tuplas a listas y viceversa. Esto lo realizamos a partir de las funciones **list()** y **tuple()** respectivamente.

```python
# Convertir una tupla a lista
mi_lista = list(mi_tupla)
print(mi_lista)
```

Al igual que con las listas, podemos conocer el número de elementos **(len)** y la posición de un elemento **(index)**.

```python
#Número de elementos
print(len(mi_tupla))

#Posición de un elemento
print(mi_tupla.index(10))
```

```python
lista1 = [1,2,3]
tuple1 = tuple(lista1)

print(tuple1)
```

### Diccionarios

Son otra manera de almacenar data. Cada valor almacenado dentro del diccionario estará relacionado con una llave. Una llave y su valor forman una pareja **“key-value”**. Los diccionarios se definen a partir de llaves **{}**.

```python
#Crear un diccionario
mi_diccionario = {'k1':'gato','k2':'perro','k3':'pez','k4':'hamster', 'k10':[1,2,3]}

#Acceder a un elemento del diccionario mediante su llave
print(mi_diccionario)
print(mi_diccionario['k1'])
print(mi_diccionario['k2'])
print(mi_diccionario['k10'])
```

Los diccionarios son objetos mutables. Se les pueden añadir yreemplazar elementos

```python
#Añadir elementos
datos = {'k5':'pájaro','k6':'toro'}
mi_diccionario.update(datos)
mi_diccionario['k6'] = 'iguana'

#Reemplazar un elemento
mi_diccionario['k1'] = 'serpiente'
print(mi_diccionario)
```

Un diccionario puede contener una lista dentro de sus elementos.

```python
#Una lista dentro de un diccionario
mi_diccionario2 = {'llave1':'Rodrigo', 'llave2':['Maria', 'Luisa']}
mi_diccionario2
```

Existen diferentes formar de crear un diccionario, una de ellas es ir asignando valores a cada llave por separado.

```python
#Forma alternativa de construir un formulario
Equipo = {}
print(type(Equipo))

Equipo['Portero'] = 'Jorge'
Equipo['Defensa'] = 'Arturo'
Equipo['Delantero'] = 'Antonio'

print(Equipo)
```
