## Gráfico de barras
<p>Los gráficos de barras visualizan datos que están organizados por categorías, donde la altura de cada barra representa los valores o la proporción de los datos en esta categoría.

En este ejercicio, tenemos la tabla <code>medals</code> la cual contiene los países y las medallas ganadas por cada uno (oro, plata y bronce).</p>

```python
# Importa Pandas y Matplotlib

import pandas as pd
import matplotlib.pyplot as plt


# Importa el archivo medals.csv a un DataFrame

medals = pd.read_csv('Data (1)/medals.csv')

medals
```

```python
# Crea la figura y el eje

fig, ax = plt.subplots()

# Traza una gráfica de barras para visualizar las medallas de oro por país

ax.bar(medals['Country'], medals['Gold'])

# Agrega etiquetas con una rotacion de 90° para los países

ax.set_xticklabels(medals['Country'], rotation=90)

# Agrega la leyenda al eje Y de "Number of medals"

ax.set_ylabel('Numero de Medallas')

# Muestra la gráfica

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,7))

# Traza una gráfica de barras para las medallas de oro por país, con la etiqueta 'Gold'

ax.bar(medals['Country'], medals['Gold'], label='Oro')

# Apila las medallas de plata a la misma grafica con la etiqueta 'Silver'

ax.bar(medals['Country'], medals['Silver'], bottom=medals['Gold'], label='Plata')

# Apila las medallas de bronce con la etiqueta 'Bronze'

ax.bar(medals['Country'], medals['Bronze'], bottom=medals['Gold']+medals['Silver'], label='Bronce')

# Agrega la leyenda a las variables

ax.legend()

# Formato chido
ax.set_ylabel('Numero de Medallas')
ax.set_xticklabels(medals['Country'], rotation=60, fontsize=10)
ax.set_title('Medallas clasificadas por pais', fontsize=20)

plt.show()
```

## Histogramas
<p>Los histogramas muestran la distribución completa de una variable. En este ejercicio, mostraremos la distribución del peso de los medallistas en gimnasia y remo en los Juegos Olímpicos de 2016 para una comparación entre ellos. La tabla que estaremos trabajando es <code>olympic2016</code></p>

```python
# Importa el archivo olympic.csv

olympic = pd.read_csv('Data (1)/olympics_2016.csv')
olympic

# Crea dos DataFrames, uno con los datos de los medallistas de gimnasia ('gimnastyc') y el otro con los medallistas de remo ('Rowing') del sexo masculino

mens_gym = olympic[(olympic['Sport'] == 'Gymnastics') & (olympic['Sex'] == 'M')]
display(mens_gym)

mens_row = olympic[(olympic['Sport'] == 'Rowing') & (olympic['Sex'] == 'M')]
display(mens_row)
```

```python
fig, ax = plt.subplots()

# Grafica un histograma para el peso de los medallistas de remo, agrega la etiqueta "Rowing" al grafico y configurado a 10 bins

ax.hist(mens_row['Weight'], label='Remo', color='red', bins=10, alpha=0.8)

# Grafica un histograma para el peso de los medallistas de gimnasia, agrega la etiqueta "Gymastics" al grafico y configurado a 10 bins

ax.hist(mens_gym['Weight'], label='Gimnasia', color='green', bins=10, alpha=0.8)

# Agrega la etiqueta al eje X: "Weight (kg)"

ax.set_xlabel('Weight (kg)')

# Agrega la etiqueta al eje Y:"# of observations"

ax.set_ylabel('# of observations')


ax.legend()


plt.show()
```

## Boxplots
<p>Otra alternatva para visualizar la distribución de una variable es mediante una gráfica de caja o <i>boxplot</i>, ahora compararemos la altura de las medallistas de Rusia y las medallistas de EE.UU.</p>

```python
fig, ax = plt.subplots()

# Crea los DataFrames para las medallistas de Rusia (RUS) y EE.UU (USA)

rus = olympic[olympic['NOC'] == 'RUS']
usa = olympic[olympic['NOC'] == 'USA']
ger = olympic[olympic['NOC'] == 'GER']
mex = olympic[olympic['NOC'] == 'MEX']

# Grafica un boxplot para comparar las alturas de las medallistas

ax.boxplot([rus['Height'], usa['Height'], ger['Height'], mex['Height']])

# Agrega las etiquetas a los datos del eje X: Russia y USA

ax.set_xticklabels(['RUS', 'USA', 'GER', 'MEX'])
ax.set_xlabel('Pais')
ax.set_ylabel('Altura (cm)')
# Agrega una eqtiqueta al eje Y: "Height (cm)"


plt.show()
```

## Graficos de dispersión
<p>La dispersión (<i>scatter</i>) es una técnica de visualización bivariada, cada registro en los datos se puede ver como un punto. La ubicación de cada punto está determinada por el valor de dos variables: la primera variable determina la distancia a lo largo del eje x y la segunda variable determina la altura a lo largo del eje y.

En este ejercicio, crearás un gráfico de dispersión de los datos de cambio climático. Regresamos a la tabla <code>climate_change</code>, la cual tiene una columna <code>co2</code> que indica las mediciones de dióxido de carbono cada mes y otra columna, <code>relativa_temp</code> que indica la temperatura media.</p>

```python
climate_change = pd.read_csv('Data (1)/climate_change.csv')


# Crear figura
fig, ax = plt.subplots()

# Gráfica una dispersión plot para visualizar la relación entre del CO2 vs la temperatura

ax.scatter(x=climate_change['co2'], y=climate_change['relative_temp'], color='orange')

# Agrega una etiqueta al eje X: "CO2 (ppm)"

ax.set_xlabel('CO2 (ppm)')

# Agrega una etiqueta al eje Y: "Relative temperature (C)"

ax.set_ylabel('Relative temperature (C)')

ax.set_title('CO2 vs Relative Temperature', fontsize=16)

plt.show()
```

```python
climate_change['date'] = pd.to_datetime(climate_change['date'])

# Crear figura

fig, ax = plt.subplots()

# Gráfica un scatter plot para visualizar la relación entre del CO2 vs la temperatura, agrega la fecha como escala de color

ax.scatter(climate_change['co2'], climate_change['relative_temp'], c=climate_change['date'])

# Agrega una eqtiqueta al eje X: "CO2 (ppm)"

ax.set_xlabel('CO2 (ppm)')

# Agrega una eqtiqueta al eje Y: "Relative temperature (C)"

ax.set_ylabel('Relative temperature (C)')

ax.set_title('CO2 vs Relative Temperature in the time', fontsize=15)


plt.show()
```
