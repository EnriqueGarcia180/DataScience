## Interfaz Matplotlib
<p>Hay muchas formas de utilizar Matplotlib. En este curso, nos centraremos en la interfaz de <code>pyplot</code>, que proporciona la mayor flexibilidad para crear y personalizar visualizaciones de datos.

Vamos a crear dos tipos de objetos: objetos de <code>Figure</code> y <code>Axes</code>. </p>

```python
# Importa pyplot de la librería Matplotlib

import matplotlib.pyplot as plt

# Crea una Figura y un Eje con el método plt.subplots

fig, ax = plt.subplots()

# Muestra el resultado
plt.show()
```

## Agregando datos al gráfico
<p>Tenemos dos tablas, la primera <code>seattle_weather</code> almacena información sobre el clima en Seattle y la segunda <code>austin_weather</code> almacena información sobre el clima en Austin. Cada una contiene una columna llamada <code>MONTHS</code> que almacena el nombre A tres letras por mes y también tiene una columna llamada <code>MLY-PRCP-NORMAL</code> que almacena el promedio de lluvia en cada mes durante un período de diez años.

Vas a crear una visualización que te permitirá comparar la precipitación en estas dos ciudades.</p>

```python
# Importa pandas

import pandas as pd

# Importa las tablas seattle_weather y austin_weather
seattle_weather = pd.read_csv('Data (1)/seattle_weather.csv')
austin_weather = pd.read_csv('Data (1)/austin_weather.csv')

display(seattle_weather[['MLY-PRCP-NORMAL', 'MONTH']])
display(austin_weather[['MLY-PRCP-NORMAL', 'MONTH']])
```

```python
# Crea una Figura y un Eje

fig, ax = plt.subplots()

# Gráfica el promedio de precipitación ('MLY-PRCP-NORMAL') de Seattle a traves de los meses ('MONTH')
# ax.plot(x data, y data)
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'])


# Gráfica el promedio de precipitación ('MLY-PRCP-NORMAL') de Austin a traves de los meses ('MONTH')
ax.plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'])

# Muestra el gráfico
plt.show()
```

## Personalizando un gráfico
<p>En este ejercicio, personalizarás la apariencia de los marcadores, el estilo de línea que se utiliza y el color de las líneas y marcadores para el gráfico.</p>

```python
fig, ax = plt.subplots()

# Gráfica la precipitación promedio de Seattle, define la línea en color rojo, con un estilo '--' y  un marcador en 'o'
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], color='r', linestyle='--', marker='o', mouseover=False)


# Gráfica la precipitación promedio de Austin, define la línea en color azul, con un estilo '--' y  un marcador en 'v'
ax.plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], color='b', linestyle='--', marker='v', mouseover=False)

# Muestra el gráfico
#plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], color='r', linestyle='--', marker='o', mouseover=False)
ax.plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], color='b', linestyle='--', marker='v', mouseover=False)

# Define la etiqueta "Time (months)" al eje X
ax.set_xlabel('Time (months)')

# Define la etiqueta "Precipitation (inches)" al eje Y
ax.set_ylabel('Precipitation (inches)')

# Agrega el titulo "Weather patterns Austin vs Seattle".
ax.set_title('Weather patterns Austin vs Seattle')

# Muestra el gráfico
plt.show()
```

## Multiples gráficos - subplots 
<p>En <b>Matplotlib</b>, se pueden crear múltiples gráficos usando la función <code>plt.subplots()</code>. El primer argumento es el número de filas en la matriz y el segundo argumento es el número de columnas. Recordemos que las tablas <code>seattle_weather</code> y <code>austin_weather</code> tienen una columna <code>MONTH</code> y las columnas <code>MLY-PRCP-NORMAL</code> (para precipitación promedio), <code>MLY-TAVG-NORMAL</code> (para temperatura promedio). En este ejercicio, graficarás la precipitación promedio mensual y las temperaturas promedio en cada ciudad en una sola vista.</p>

```python
# Crea una figura para colocar subplots en 2 filas y 2 columnas

fig, ax = plt.subplots(2, 2, figsize=(12,10))

# En el eje superior izquierdo (index 0, 0), grafica la precipitación mensual de Seattle
ax[0,0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'])
ax[0,0].set_title('Precipitacion Mensual - Seattle')
ax[0,0].set_xlabel('Month')
ax[0,0].set_ylabel('Precipitation (inches)')

# En el eje superior derecho (index 0, 1), grafica la temperatura promedio mensual de Seattle
ax[0,1].plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'])
ax[0,1].set_title('Temperatura Mensual - Seattle')
ax[0,1].set_xlabel('Month')
ax[0,1].set_ylabel('Temperature (fahrenheit)')

# En el eje inferior izquierdo (1, 0) grafica la precipitación mensual de Austin
ax[1,0].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'])
ax[1,0].set_title('Precipitacion Mensual - Austin')
ax[1,0].set_xlabel('Month')
ax[1,0].set_ylabel('Precipitation (inches)')

# En el eje inferior derecho (1, 1) grafica la temperatura promedio mensual de Austin
ax[1,1].plot(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'])
ax[1,1].set_title('Temperatura Mensual - Austin')
ax[1,1].set_xlabel('Month')
ax[1,1].set_ylabel('Temperature (fahrenheit)')

plt.show()
```

## Usando la misma escala
<p>Al crear múltiples gráficos, a menudo es preferible asegurarse de que estos se muestren con la misma escala utilizada en el eje Y. Esto se puede configurar estableciendo la palabra clave <code>sharey=True</code>.</p>

```python
# Crea una figura con ejes de 2 filas, 1 columna y el eje Y compartido para la misma escala

fig, ax = plt.subplots(2,1, figsize=(10,7), sharey=True)

# Gráfica la precipitación normal de Seattle "MLY-PRCP-NORMAL" con una línea color rojo, además agrega los tipos de precipitaciones "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL" con líneas punteadas color azul y verde
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], color='r')
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-25PCTL'], color='b', linestyle='--')
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-75PCTL'], color='g', linestyle='--')
ax[0].set_title('seattle weather')
ax[0].legend(['norm', '25', '75'], loc='lower left')

# Gráfica la precipitación normal de Austin "MLY-PRCP-NORMAL" con una línea color rojo, además agrega los tipos de precipitaciones "MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL" con líneas punteada color azul y verde
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], color='r')
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-25PCTL'], color='b', linestyle='--')
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-75PCTL'], color='g', linestyle='--')
ax[1].set_title('austin weather')
ax[1].legend(['norm', '25', '75'], loc='upper right')

plt.show()
```
