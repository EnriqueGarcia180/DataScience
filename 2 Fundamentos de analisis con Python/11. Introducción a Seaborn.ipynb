## Comparando histogramas
<p>Vamos a analizar datos de los programas que el gobierno de EE.UU proporciona a las escuelas como presupuesto para infraestructura. La librería de <code>seaborn</code> proporciona herramientas de visualización para análsis estadistico mucho más sofisticado. amos a comparar la diferencia con otras librerías </p>

```python
# Importar pandas, seaborn, y matplotlib.pyplot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Importa el archivo US School Improvement Grants 2010.csv
df = pd.read_csv('Data (1)/US School Improvement Grants 2010.csv')
df
```

```python
# Grafica un histograma del monto de los premios 'Award_amount' con *** pandas ***
df['Award_Amount'].plot.hist()
plt.show()
```

```python
# Grafica un histograma del monto de los premios 'Award_amount' con *** seaborn ***
sns.displot(df['Award_Amount'])
plt.show()
```

## distplot
<p>La función <code>distplot()</code> nos muestra la <a href="https://en.wikipedia.org/wiki/Kernel_density_estimation">Densidad Estimada Kernel (KDE)</a>, la cual a grandes rasgos es una forma de suavizar la curva de distribución de una variable </p>

```python
# Crea un displot de 'Award_amount' con la función KDE desactivada y con 20 bins

sns.displot(df['Award_Amount'], bins=20)


plt.show()
```

```python
# Crea un displot de 'Award_amount' con la función hist desactivada y las funciones rug y shade activas

sns.displot(df['Award_Amount'], kind='kde', fill=True, rug=True)
# Resultado
plt.show()
```

## Graficos de Regresión
<p>Para este conjunto de ejercicios, analizaremos los datos de FiveThirtyEight sobre qué estado de EE.UU. Tiene la mayor tasa de siniestros en accidentes de automovil. El dataset incluye información resumida sobre accidentes mortales, así como primas de seguro para cada estado a partir de 2010.</p>

```python
# Importa el archivo Insurance Premiums Automobile.csv

insurance = pd.read_csv('Data (1)/insurance_premiums.csv')
insurance
```

```python
# Crea una grafica de regresion para visualizar la relación que existe en las primas (premiums) y las perdidas de seguro (insurance_losses)

sns.regplot(data=insurance, x=insurance['insurance_losses'], y=insurance['premiums'])

plt.show()

# el canal sombreado muestra el rango en el que se puede mover la linea
```

```python
# Con la funcion ** lmplot ** gráfica la relación de las primas (premiums) y las perdidas de seguro (insurance_losses) por Region (Representada en colores)

sns.lmplot(data=insurance, x='insurance_losses', y='premiums', hue='Region')

plt.show()
```

```python
# Con la funcion lmplot gráfica la relación de las primas (premiums) y las perdidas de seguro (insurance_losses) por Region (Representada en columnas)

sns.lmplot(data=insurance, x='insurance_losses', y='premiums', col='Region')

plt.show()
```

```python
# Con la funcion lmplot gráfica la relación de las primas (premiums) y las perdidas de seguro (insurance_losses) por Region (Representada en filas)
sns.lmplot(data=insurance, x='insurance_losses', y='premiums', row='Region')

plt.show()
```
