## 1. Generación de números pseudoaleatorios

La librería numpy contiene el módulo **numpy.random**, el cuál genera números pseudo-aleatorios a partir de una variedad de distribuciones de probabilidad. 
A partir de los procesos contenidos en este módulo, es posible crear muestras de números aleatorios a partir de diferentes métodos.

El proceso general para producir muestras de números aleatorios es **default_rgn**.

```python
#Importamos numpy

import numpy as np

# Generador de números aleatorios (utilizando una SEMILLA de 150)

# rng: random number generator

rng = np.random.default_rng(150)
print('rng', rng)


#Utilizamos el generador para crear números

muestra = rng.random(10)
print('muestra',muestra)


# sin semilla, numpy selecciona una semilla por si solo

rng1 = np.random.default_rng()
print('rng1',rng1)

muestra2 = rng1.random(10)
print('muestra2',muestra2)
```

Una de las maneras de generar números pseudo-aleatorios es siguiendo una distribución normal. Para esto, utilizamos **np.random.standard_normal**.

```python
#Muestra de 4x4 siguiendo la distribución normal

muestra2 = np.random.standard_normal(size = (4,4))
print('muestra2', muestra2)
```

### Estadística con Numpy

Numpy contiene una amplia variedad de funciones matemáticas dentro de las cuáles se destacan las funciones estadísticas. A partir de estas funciones, podemos conocer las estadísticas principales de un set de datos como lo son la **media**, **varianza**, **desviación estándar**, **máximo**, **mínimo**, etc.

```python
#Creamos una muestra de 100 datos que siguen una distribución normal
# mu = 5, sigma = 2

mu = 5
sigma = 2

datos = np.random.normal(mu, sigma, size=1000)
print('datos', datos)
```

```python
#Obtenemos las principales estadísticas

#media (arriba establecimos mu = 5)
print('media',datos.mean())

#varianza (arriba dijimos sigma = 2 entonces sigma^2 ~ 4)
print('varianza',datos.var())

#desv est (arriba dijimos sigma = 2)
print('desv estandar', datos.std())

#min
print('min',datos.min())

#max
print('max',datos.max())
```

### La distribución normal

A partir de la muestra de números aleatorios que siguen una distribución normal, podemos estudiar tanto estadística como gráficamente esta distribución de probabilidad, la cúal siguen muchos fenómenos físicos, biológicos y económicos.

```python
# histograma de la distribución normal

import matplotlib.pyplot as plt

bins = 30 #numero de barras para el histograma

plt.hist(datos, bins=30, color='blue', alpha=0.7)
plt.title('Histograma de la variable "datos" ')
plt.ylabel('Frequency')
plt.xlabel(' "datos" ')
plt.show()
```

## Ejemplo: caminatas aleatorias (Random Walk).
 
Proceso Estocastico, el mas sencillo

```python
posicion = 0

caminata = [posicion]
n_pasos = 100

# moneda al aire lanzada 100 veces, ver ganancias acumuladas
for i in range(n_pasos):
    if np.random.randint(0,2) == 1:
        paso = 1
    else:
        paso = -1
    posicion += paso
    caminata.append(posicion)

caminata_array = np.array(caminata)
print('Ultimo valor:', caminata_array[100])
print('Mean', caminata_array.mean())
print('Std Dev:', caminata_array.std())
print('Min:', caminata_array.min())
print('Max:', caminata_array.max())

plt.plot(caminata)
plt.show()
```
