## Estilos de grafica
<p>Para estos ejercicios, analizaremos los valores de renta de mercado calculados por el Departamento de Vivienda y Desarrollo Urbano de EE.UU. Estos datos se utilizan para calcular los estandares para varios programas federales, los valores reales de los alquileres lllegan a varíar mucho en los EE. UU </p>

```python
# Importar pandas, seaborn, y matplotlib.pyplot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importa el archivo US Housing.csv

df = pd.read_csv('Data (1)/US Housing.csv')
df
```

```python
# Con este comando borramos las configuraciones existentes en nuestra grafica
plt.clf()

# Grafica un histograma con pandas visualizar la distribución del costo del alquiler de un apartamento de 2 habitaciones 'fmr_2'
df['fmr_2'].plot.hist()
plt.show()
plt.clf()

# # Aplica la función para estilo seaborn
sns.set()

# Grafica el histograma nuevamente pero ahora con seaborn
df['fmr_2'].plot.hist()
plt.show()
plt.clf()
```

```python
# Grafica la misma variable del ejercicio anterior pero con un estilo 'whitegrid'

sns.set_style('whitegrid')

sns.displot(df['fmr_2'])

plt.show()
```

```python
# Grafica la misma variable del ejercicio anterior pero con un estilo 'dark'
sns.set_style('dark')
sns.displot(df['fmr_2'])
plt.show()
plt.clf()
```

```python
# Define un estilo 'white'

sns.set_style('white')

# Crea una grafica de regresion (lmplot) para poder visualizar las personas que alquilan un departamento de 2 habitaciones (pop2010 vs fmr_2)
sns.lmplot(data=df, x='pop2010', y='fmr_2')

# Muestra la grafica y borra la configuración
plt.show()
plt.clf()
```

## Configurando el color
<p>Seaborn ofrece varias opciones para modificar los colores de nuestras visualizaciones.</p>

```python
# Habilita los codigos de color y grafica una curva de distribucion para las personas que alquilan deportamentos de 3 habitaciones en color magenta (m)
sns.set_style('whitegrid')
sns.displot(df['fmr_3'], color='m')
plt.show()
plt.clf()
```

```python
# Visualiza la distribución anterior entre las paletas de color 'bright', 'colorblind' y 'muted'
for p in ['bright', 'colorblind']:
    sns.set_palette(p)
    sns.displot(df['fmr_3'])
    plt.show()
    plt.clf()
```

```python
# Despliega una paleta Purpura con una escala de 8 colores (Purples)

sns.palplot(sns.color_palette('Purples', 10))

plt.show()
```

```python
# Despliega una paleta de Azul a Rojo con una escala de 12 colores (coolwarm)

sns.palplot(sns.color_palette('coolwarm', 12))
plt.show()
```

## Matplotlib + Seaborn
<p>La mayoría de las veces podemos usar la librería de <code>seaborn</code> para graficar nuestras visualizaciones, pero a veces es útil usar las funciones de <code>matplotlib</code> para dar una mayor personalización</p>

```python
# Crea la figura y el eje
fig, ax = plt.subplots()

# Grafica la distribución de renta de apartamentos de 3 habitaciones

sns.histplot(df['fmr_3'])

# Agrega la leyenda al eje X: "3 Bedroom Fair Market Rent"

ax.set_xlabel('3 Bedroom Fair Market Rent')

plt.show()
```

```python
# Crea la figura y el eje

fig, ax = plt.subplots()

# Grafica la distribución de la renta de apartamentos de 1 habitación

sns.histplot(df['fmr_1'])

# Agrega la leyenda al eje X: "1 Bedroom Fair Market Rent", define un limite entre 500 y 1500 y agrega un titulo a la grafica: "US Rent"

ax.set(xlabel='1 Bedroom Fair Market Rent', xlim=(500, 1500), title='US Rent')

plt.show()
```

```python
# Calcula la media y la mediana de la renta de apartamentos de 1 habitación
import numpy as np
mean = np.mean(df['fmr_1'])
# without numpy:
median = df['fmr_1'].median()

print(mean)
print(median)

# Crea el mismo grafico del ejercicio anterior
fig, ax = plt.subplots()
sns.histplot(df['fmr_1'])


# Pero antes de visualizarlo vamos a agregar algunas anotaciones, coloca los valores de la media y la mediana
ax.axvline(x=median, color='m', label='Mediana', linestyle='--', linewidth=2)
ax.axvline(x=mean, color='b', label='Mean', linestyle='-', linewidth=2)

# Agrega la leyenda a los ejes
ax.legend()

plt.show()
```

<p>
    

Ahora vamos a comparar el costo de la renta de los departamentos de 1 habitación y 2 habitaciones</p>

```python
# Crea una figura con dos ejes, 1 fila y dos columnas con eje compartido

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

# Grafica en el eje 1 la distribucion de la renta de departamentos de 1 habitación, agrega una leyenda al eje X para distinguir y toma en cuenta un rango entre 100 y 1500

sns.histplot(df['fmr_1'], ax=ax1)
ax1.set(xlabel='Viviendas de 1 Habitacion', xlim=(100, 1500))

# Grafica en el eje 2 la distribucion de la renta de departamentos de 1 habitación, agrega una leyenda al eje X para distinguir y toma en cuenta un rango entre 100 y 1500

sns.histplot(df['fmr_2'], ax=ax2)
ax2.set(xlabel='Viviendas de 2 Habitaciones', xlim=(100, 1500))

plt.show()
```
