# Regresion Lineal - Prediccion

$\qquad\qquad  y=mx+b$

$\qquad\qquad  y = w_1x_1+w_2x_2+...+w_nx_n+b$

$\qquad\qquad  \text{MSE (mean squared error) to adjust weights (w's) or coeficients:}$

$\qquad\qquad  Error = \sum_{}^{} (y_{real} - y_{predicted})^2 $


**Ejercicio**:  Predecir el precio de un a casa en funcion de su superficie en m2

```python
#importar librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Datos de Entrenamiento

```python
#Datos de ejemplos sobre superficio en m2 y precio en miles de dolares

x = np.array([[50],[60],[80],[100],[120],[150]]) #variable independiente: superficie en m2 de las casas
y = np.array([100, 120, 160, 200, 240, 300]) #variable dependiente: precio
```

Entrenamiento

```python
#Crear y Entrenar el modelo de Regresion Lineal
modelo = LinearRegression()
modelo.fit(x,y) #precio = m*superficie + b
```

Nuevos Valores para Predecir

```python
x_nuevo = np.array([[90], [110], [130]])
```

Predicciones

```python
y_pred = modelo.predict(x_nuevo)
```

```python
#imprimir coeficientes
print("Pendiente (coeficiente): ",modelo.coef_[0]) #slope de cuanto aumenta el precio por cada m2
print("Interseccion (bias): ",modelo.intercept_) #precio base cuando la superficie es "0"
print("Predicciones: ",y_pred)
```

Graficas

```python
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, modelo.predict(x), color='red', label='Linea de regresion')
plt.scatter(x_nuevo, y_pred, color='green', label='Predicciones nuevas')
plt.xlabel('Superficie en m2')
plt.ylabel('Precio miles-usd $')
plt.title('Regresion Lineal Precio vs Superficie')
plt.legend()
plt.grid(True)
plt.show()
```
