# Regresion Logistica - Clasificacion

Es un modelo Parametrico, aprende una funcion matematica (sigmoid function) que separa las clases

- Clasificar correos como spam o no spam (clasificacion binaria)
- Predecir si un cliente compra o no compra
- Determinar si un paciente tiene o no una enfermedad
- Clasificacion multiclase



## Clasificacion Binaria

$\text{Sigmoid function:}$

$\qquad\qquad  P(y=1)=\frac{1}{1+ \exp^{-(mx+b)}}$

$\qquad\qquad  \text{Si P >= 0.5 , clase 1}$

$\qquad\qquad  \text{Si P < 0.5 , clase 0}$



* Rango: el resultado siempre esta entre 0 y 1
* Si z -> +inf, S(z) -> 1
* Si z -> -inf, S(z) -> 0
* Forma de S
* Valor en 0: S(0)=0.5

En clasificacion binaria, la salida de sigmoide se interpreta como la probabilidad de pertenecer a la clase 1

Ejemplo:  

Si S(z)=0.9 -> 90% de probabilidad de ser clase 1

Luego se aplica un **umbral** (default 0.5)

Si >= 0.5 -> clase 1

Si < 0.5 -> clase 0

## Sigmoid function:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from IPython.display import display, Math
display(Math(r"S(z) = \frac{1}{1+ \exp^{-z}}"))
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
#valores para graficar
z = np.linspace(-10, 10, 100)
s = sigmoid(z)
```

```python
#graficar
plt.plot(z,s)
plt.title("Sigmoid function")
plt.xlabel("z")
plt.ylabel('S(z)')
plt.grid(True)
plt.axhline(0.5, color="red", linestyle="--")
plt.show()
```
