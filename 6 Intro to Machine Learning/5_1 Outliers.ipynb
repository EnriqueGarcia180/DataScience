```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
```

```python
datos = np.random.normal(loc=1.65, scale=0.05, size=100)
```

```python
# Creamos outliers
outliers = np.array([1.30, 1.90, 2.10, 1.20])

# Agregamos outliers a datos
datos = np.concatenate([datos, outliers])
```

```python
#Grafica Boxplot con matplotlib
plt.figure(figsize=(8,4))
plt.boxplot(x=datos)
plt.title("Deteccion de Outliers en base de datos")

#Grafica Boxplot con seaborn
plt.figure(figsize=(8,4))
sns.boxplot(x=datos, color='Skyblue')
plt.title("Deteccion de Outliers en base de datos")
```

```python
# Grafica Scatterplot
plt.figure(figsize=(8,4))
plt.scatter(x=range(len(datos)), y=datos, color="blue", s=60, label="Datos")
plt.axhline(np.mean(datos), color="red", linestyle="--", label="Media")
plt.title("Visualizacion de outliers (Scatter plot)")
plt.xlabel("Indice")
plt.ylabel("Altura (m)")
plt.legend()
plt.show()
```

```python
# Metodo IQR para detectar Outliers

Q1 = np.percentile(datos, 25) #25% de los datos caen en el primer percentil o quartil
Q3 = np.percentile(datos, 75)

IQR = Q3 - Q1 #mide la dispersion de los datos o rango inerquartl
```

```python
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
```

La regla clasica para detectar los outliers:  
Todo valor < Limite Inferior o > Limite Superior es un outlier

```python
outliers_detectados = datos[(datos < limite_inferior) | (datos > limite_superior)]
print("outliers_detectados:", outliers_detectados)
```
