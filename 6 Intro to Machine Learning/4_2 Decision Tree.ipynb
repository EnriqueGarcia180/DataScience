# Decision Tree - Prediccion

Introducción a Random Forest y árboles de decisión.

Entrenamiento con diferentes subconjuntos de datos.

Precisión alta en entrenamiento, riesgo de sobreajuste.

Ventajas: robustez y análisis de variables importantes.

Desventajas: mayor complejidad y menor interpretabilidad.

Ejercicio: Predecir si un alumno aprueba un examen segun sus horas de estudio y horas de sueño

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Datos
x = np.array([[2,5], [4,6], [6,7], [8,4], [10,3]]) #[horas de estudio, horas de sueño]
y = np.array([0,0,1,1,1]) # 0=No aprueba, 1=Aprueba
```

```python
# Crear y entrenar el Modelo
arbol = DecisionTreeClassifier(criterion='gini', max_depth=3)
arbol.fit(x, y)
```

```python
#visualizaicón
plt.figure(figsize=(6,6))
plot_tree(arbol, feature_names=["Horas estudio", "Horas sueño"], class_names=["No aprobo", "Sí aprobo"], filled=True)
plt.show()
```

Conclusion: El arbol predice que si una persona estudia mas de 5 horas SI va a aprobar. EL modelo infirio que la variable Horas de Estudio tiene mayor peso en el resultado.

Comprobacion con una nueva persona:

```python
# Prediccion de un nuevo caso
nueva_persona = np.array([[7,6]])
print("Prediccion:", arbol.predict(nueva_persona))
# 0=No aprueba, 1=Aprueba
```
