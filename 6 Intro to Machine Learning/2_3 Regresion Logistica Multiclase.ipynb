## Regresion Multiclase - Clasificacion


    
    




```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
x = np.array([   # variables independientes:  peso, color
    [150, 0],
    [180, 1],
    [120, 2],
    [130, 2],
    [200, 1]
])
```

```python
y = np.array(["Manzana", "Manzana", "Platano", "Platano", "Cereza"])
```

```python
modelo = LogisticRegression(multi_class='multinomial', solver='lbfgs')
modelo.fit(x,y)
```

```python
#dato nuevo
x_nuevo = np.array([[160, 0]])
```

```python
#prediccion
print("Prediccion con dato nuevo: \n",modelo.predict(x_nuevo)) 
print("Mostrar las Clases del modelo: \n",modelo.classes_)
print("Probabilidad del nuevo dato de pertenecer a cada una de las clases del modelo: \n",modelo.predict_proba(x_nuevo)) 
```
