# Regresion Logistica - Clasificacion


**Ejercicio:** Determinar si con 4.5 horas de estudio un estudiante aprobara el examen on un 70% de probabilidad de pasar

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
x = np.array([[1],[2],[3],[4],[5],[6]]) # horas de estudio
y = np.array([0,0,0,1,1,1]) # si aprobo = 1, no aprobo = 0
```

```python
#crear y entrenar el modelo
modelo = LogisticRegression()
modelo.fit(x,y)
```

```python
#Prediccion con 300 datos de prueba
x_test = np.linspace(0,7,300).reshape(-1,1) #crea 300 valores entre 0 y 7 horas de estudio y lo combierte a un arreglo columnas
y_prob = modelo.predict_proba(x_test) #calcula la probabilidad de pertenencia a la clase 1
```

```python
#Prediccion para un caso especifico: 4.5hrs de estudio para pasar on un 70% de probabilidad esperada
nueva_hora = np.array([[4.5]])
probabilidad_aprobar = modelo.predict_proba(nueva_hora)[0,1] # para que nos de la probabilidad de aprobar (renglon 0, columna 1 (aprobar)), columna 0 es no aprobar
print(f"probabilidad_aprobar: {100*probabilidad_aprobar:.2f}%")
```

Graficar la Regresion Logistics de los 300 datos de prueba asi como el caso especifico de 4.5hrs de estudio:

```python
plt.Figure(figsize=(8,6))
plt.scatter(x,y, color='Blue', label="Datos Reales")
plt.plot(x_test, y_prob[:,0], color='red', label = 'Curva Logistica probabilidad de ser 0')
plt.plot(x_test, y_prob[:,1], color='green', label = 'Curva Logistica probabilidad de ser 1')
plt.axhline(0.5, color='gray', linestyle='--', label='Umbral 0.5')
plt.scatter(nueva_hora, probabilidad_aprobar, color="orange", label=f'Prediccion (4.5h):{probabilidad_aprobar:.2f}')
plt.xlabel('Horas de Estudio')
plt.ylabel('Probabilidad de Aprobar')
plt.title('Regresion Logistica: Horas de Estudio vs Aprobacion')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
