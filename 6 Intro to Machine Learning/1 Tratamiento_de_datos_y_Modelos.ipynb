# Tratamiento de un Modelo

1. Recoleccion de datos
2. Procesamiento de datos
   1. Eliminar valores faltantes
   2. Eliminar errores
   3. Codificar (Encoder, de variables categoricas a numericas)
   4. Normalizar datos (que esten en la misma escala)
3. Seleccion del modelo (regresion, Clasificacion, Clustering)
4. Entrenamiento del modelo
5. Evaluacion del modelo
   1. Precision, Recall, F1-Score para clasificacion
   2. Error cuadratico Medio para regresion
6. Prediccion

```python
# Cargar librerias
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

## Ingresar datos manualmente en DataFrame

```python
# Cargar datos
data = pd.DataFrame({
    'Nombre': ['Ana','Luis','Martha','Carlos'],
    'Edad': [18,17,19,None],
    'Genero':['F','M','F', 'M'],
    'Nota1':[7.0, None, 6.0, 5.5],
    'Nota2':[8.0, 6.5, 7.0, None],
    'Asistencia':[0.9, 0.8, 0.95, 0.75],
    'Aprobado':['Si','No','Si','No']
})
print(data)
```

```python
# Reemplazar los None con la media de la columna
data['Edad'].fillna(data['Edad'].mean(), inplace=True)
data['Nota1'].fillna(data['Nota1'].mean(), inplace=True)
data['Nota2'].fillna(data['Nota2'].mean(), inplace=True)
```

```python
# Codificar (Encoder) 'Genero' y 'Aprobado', es decir convertir de variable categorica (texto) a numero
le = LabelEncoder()
data['Genero'] = le.fit_transform(data['Genero'])
data['Aprobado'] = le.fit_transform(data['Aprobado'])
```

```python
# Normalizar 'Edad', 'Nota1', 'Nota2' y 'Asistencia'
scaler = StandardScaler()
data[['Edad','Nota1','Nota2','Asistencia']] = scaler.fit_transform(data[['Edad','Nota1','Nota2','Asistencia']])
```

```python
print(data)
```
