```python
import yfinance as yf
yf.__version__

# values are adjusted including dividends. No "Adj Close" column appears now.
df_ = yf.download(['NVDA', 'TSLA'], start='2022-01-01')

# shows that now is multiple level index column
df_
```

```python
# reduction to only NVDA using cross section funcion. 
df_.xs('NVDA', axis=1, level='Ticker')
```

```python
# if we don't want values adjusted by dividends

df_noAdj = yf.download('NVDA', start='2022-01-01', auto_adjust=False)

# See how "Adj Close" column appears now and "Close, High, Low, Open" column is different now.
df_noAdj
```

```python
#factor asjustment example: Close / adj Close
print('Factor Adjustment:', 30.066137313842773 / 30.121000289916992)

# same for High, Low, Open.

# comprobacion: multiplicar el Adjustment factor x Low
print('Comprobacion para "Low Not Adjusted":',29.78499984741211,'x adjustment factor', 0.9981785805402823, '=',29.78499984741211*0.9981785805402823)
print('Low adjusted: ',29.730748869082344)
```
