# Z-Score



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display, Math
```

```python
# generar datos aleatorios
np.random.seed(42)
datos = np.random.normal(loc=1.65, scale=0.05, size=100)
```

```python
# outliers
outliers = np.array([1.30, 1.90, 2.10, 1.20])
datos = np.concatenate([datos, outliers])
```

```python
# formula z-socre
from IPython.display import display, Math
display(Math(r"Z = \frac{(x-\mu)}{\sigma}"))

# Calculo de z-score
media = np.mean(datos)
desviacion = np.std(datos)
z_scores = (datos - media) / desviacion
```

```python
outliers_detectados = datos[np.abs(z_scores)>3]
```

```python
# Imprimir datos
print("Media:",round(media,3))
print("Desviacion estandar:", round(desviacion, 3))
print("Z-score de los datos", np.round(z_scores, 3))
print("Outliers detectados:", outliers_detectados)
```

```python
# visualizacion
plt.figure(figsize=(8,4))
plt.scatter(x=range(len(datos)), y=z_scores, color="blue", s=60, label="Z-score")
plt.axhline(3, color="red", linestyle="--", label="Limites (+/-)3")
plt.axhline(-3, color="red", linestyle="--")
plt.title("Deteccion de outliers con Z-score")
plt.xlabel("Indice")
plt.ylabel("Z-scores")
plt.legend()
plt.show()
```

nota: el 1.90 NO lo considero como "outlier" porque su z-score fue 2.93 por lo que queda dentro del limite de +/- 3
